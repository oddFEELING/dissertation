```python
import pandas as pd

df = pd.read_csv("./human_disease_experiments_filtered.tsv", sep='\t',
                 names=['gene_ids', 'gene_name', 'disease_identifier',
                        'disease_name', 'db', 'source_score', 'confidence_score'])
df
```

<div style="background-color: #0a1a1a; color: #d3d3d3; padding: 10px; margin:10px 0; border-radius: 10px; box-shadow: 0 4px 8px rgba(0,0,0,0.2);">
    <p style="color: #e6e6e6; text-align: center; font-family: 'Arial', sans-serif; text-transform: uppercase; letter-spacing: 2px; margin: 0;">
      Drop columns not needed for analysis
    </p>
</div>

```python
### Drop columns not needed for analysis  
df.drop(['disease_identifier', 'db'], axis=1, inplace=True)
df
```

<div style="background-color: #0a1a1a; color: #d3d3d3; padding: 10px; margin:10px 0; border-radius: 10px; box-shadow: 0 4px 8px rgba(0,0,0,0.2);">
    <p style="color: #e6e6e6; text-align: center; font-family: 'Arial', sans-serif; text-transform: uppercase; letter-spacing: 2px; margin: 0;">
      Convert string source score to an integer
    </p>
</div>

```python
df['source_score'] = df['source_score'].apply(lambda x: int(x.split(' ')[2]))

df
```

<div style="background-color: #0a1a1a; color: #d3d3d3; padding: 10px; margin:10px 0; border-radius: 10px; box-shadow: 0 4px 8px rgba(0,0,0,0.2);">
    <p style="color: #e6e6e6; text-align: center; font-family: 'Arial', sans-serif; text-transform: uppercase; letter-spacing: 2px; margin: 0;">
        Drop entries with confidence score < 0.5
    </p>
</div>

```python
### drop entries with poor confidence score  
df = df[df['confidence_score'] > 0.5]
df
```

```python
from src.data.cancer_diseases import cancer_diseases

### Only keep genes that are expressed in cancerous diseases  
df = df[df['disease_name'].isin(cancer_diseases)]

df
```
